```python
import pandas as pd
```

```python
df = pd.read_csv("pokedex_markdown.csv")
df.head()
```

```python
mons = ["Charmander", "Pikachu", "Squirtle", "Bulbasaur"]
df[df["Name"].isin(mons)]
```

```python
df.set_index("Name", inplace=True)
```

```python
new_df = df.loc[mons]
new_df.reset_index(inplace=True)
new_df
```

```python
src_df = new_df[['Name', 'Image', 'Type', 'Ability', 'Hidden Ability']].copy()
src_df
```

```python
new_df2 = pd.DataFrame(columns=src_df.columns)
new_df2.loc[0] = src_df.columns
new_df2 = pd.concat([new_df2, src_df])
new_df2 = new_df2.set_index('Name').reset_index()
new_df2
```
